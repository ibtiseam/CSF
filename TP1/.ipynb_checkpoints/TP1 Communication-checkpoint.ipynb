```python
import math 
from math import *
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from dec2bin import *
import math
import staircase as sc
```

```python
f=10
fe=1000
te=1/fe
t=arange(0,1,te)
s=[sin(2*pi*f*i)for i in t]
plt.plot(t,s)
plt.show()
```

```python
N=len(t)
sff=np.fft.fftshift(np.fft.fft(s))
sf=sff/N
f=np.linspace(-fe/2,fe/2,N)
```

```python

figure=plt.figure()
axis=figure.add_subplot(111)
axis.set_xlim(-15,15)
plt.stem(f,abs(sf))
plt.show()
```

```python
fe = 1000
te = 1/fe
f=10
t=arange(0,1,te)
n = 3
```

```python
s=[sin(2*pi*f*i)for i in t]
smin=np.amin(s)
smax=np.amax(s)
pas=(smax-smin)/(2**n)
a=floor((s-smin)/pas)
sq=smin+pas*a+(pas/2)
```

```python
for i in range(n):
    if (sq[i] >= smax):
        sq[i]=smax-(pas/2)
```

```python
plt.plot(t,sq)
plt.plot(t,s)
plt.show()
```

```python
fe = 1000
te = 1/fe
f=10
t=arange(0,1,te)
n = 3
```

```python
s=[sin(2*pi*f*i)for i in t]
smin=np.amin(s)
smax=np.amax(s)
pas=(smax-smin)/(2**n)
quan=floor((s-smin)/pas)

sq=smin+pas*quan+(pas/2)

for i in range(1,len(sq)):
    if (sq[i] >= smax):
        sq[i]=smax-(pas/2)
```

```python
dig= []
for i in range(len(quan)):
    dig.append(dec2bin(quan[i]))
```

```python
figure=plt.figure(figsize=[10,10])
plt.plot(t,dig,'r')
plt.plot(t,s,'r')

plt.grid(True)
plt.show()
```
